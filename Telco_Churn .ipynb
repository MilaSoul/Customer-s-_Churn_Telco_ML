## Reading the outcomes  
Going deeper into the explanation and always seeking reliable information, code will be applied to identify even further the efficiency of the predictions. After applying the Confusion Matrix, we will be able to analyse specific indexes such as :

- **Precision:** Brings the fraction of positive predictions that were actually positive. Consider the proportion of misclassified classes, in other words, the precision of the positive samples classified. 
- **Recall:** It is the measurement of the proportion of the positive samples identified by the model. 
- **F1-Score:** A combination between both indexes cited above, mathematically a "harmonic mean of precision"(Mohajon, 2020)
- **Accuracy:** As the name suggests, the overall accuracy of the model, when the classes of the samples were predicted correctly by the classifier.

# Machine Learning model : Decision Tree

Decision Tree was picked up to be one of the machine learning models as this method aims to work not just with numerical features but with categorical ones as well. As we could see before the dataset we are running, the majority of the columns are categorical ones. The Decision Tree model will be useful for classification purposes. That model will lead us, of course according to the algorithm, to the most likely combination of observations into the core goal of the business issue: either the customer will be a churn yes or no, by splitting up the observations we have on the data set. 


One of the perks of that machine learning model is that we don't need to carry out an encoding method before running the algorithm, we can apply the Decision Tree straightway on the dataset without that procedure. 
In the following code, the decision tree will be taken randomly combination of observations by the Gini index. "The lower the Gini Index, the better the lower the likelihood of misclassification."(Dash, 2022)

*Dash, S. (2022). Decision Trees Explained — Entropy, Information Gain, Gini Index, CCP Pruning.. [online] Medium. Available at: https://towardsdatascience.com/decision-trees-explained-entropy-information-gain-gini-index-ccp-pruning-4d78070db36c.*

# Decision Tree

```python
# k fold cross validatio
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score 
import sklearn.metrics as metrics

DT = DecisionTreeClassifier(max_depth = 3, random_state = 0)# max deep for the performance is 3 
kfold = StratifiedKFold(n_splits=7, random_state=1, shuffle=True)#cross-validation by Kfold technic


DT.fit(X_train, Y_train)
y_pred_DT = DT.predict(X_test)

cvs_DT = cross_val_score(estimator = DT, X = X_train, y = Y_train, cv = kfold, scoring='accuracy')

#confusion matrix 
cm = confusion_matrix(Y_test, y_pred_DT)

# Call a method heatmap() to plot confusion matrix
sns.heatmap(cm, annot = True)

Score = metrics.accuracy_score(Y_test, y_pred_DT)
print("Cofusion matrix: " '\n', confusion_matrix(Y_test, y_pred_DT))


print('Acuracy: ',Score)
```

```python
print("Percent of accuracy: ", cvs_DT) 
print()
print("Mean Accuracy :", cvs_DT.mean(), "\nStandard Deviation :", cvs_DT.std())
print()
print("Classification Report: ", classification_report(Y_test, y_pred_DT))
```

Mean Accuracy : 0.76

The matrix displays True Positeve TP(996) and False Negative (65) have a high number of true values compared to other models. Based on precision we also can say that the model does perform well as it takes 82% and 66% of relevant values. Based on the recall we can say the model to find well 0 class but a worse 1 class


# Machine Learning model : RANDOM FOREST 

As worked previously with Decision Tree, now we are going to with a more complex Machine Learning model: Random Forest. Random Forest Classifier is a step further in the prediction based on the observations. This model aims to collect the data, and build a large number of possibilities, which we could different kinds of variations of Decision Trees and so as an outcome, show the most accurate combination among the observations collected. This model can be applied to both regression and classification problems, giving good results even without hyper parameter estimation.

The Random forest works with a training method where in each decision tree made, the following one will be impruved by the training (or we could say by the learning) and so go get towards ahead a more accurate prediction. The model will work with Feature Randomness, a normal random forest method, where it will "consider every possible feature and pick the one that produces the most separation between the observations in the node".(Yiu, 2019).

This method will aim to creat the most diverse random possibilities among the observations, avoinding correlation between them. The target variable in our model is 'Churn', we look at its distribution and separate it from the rest of the data set while running the code.

# RANDOM FOREST & Cross Validation

# Machine Learning model: LogisticRegression 
Logistic regression is a binary classification algorithm and is simply a neural network. In this sense, it forms the basis of deep learning. While linear regression aims to find the best straight line, logistic regression aims to classify data by creating the best S curve. It outputs a probability estimate.

Logistic regression estimates the probability of an event occurring, such as churn or no churn in a particular case, based on the dataset of independent features. 

This method tests different values through multiple iterations in order to find the best fit of log odds.
The model required a large dataset in order to have high statistical power to detect a significant effect.

In this case, we are working with binary regression. it has only two possible outcomes 0 or 1
The model is the most commonly used approach for this type of classification. 

IBM.2022., What is logistic regression? [online]. Available at: https://www.ibm.com/topics/logistic-regression [Acceced 25 November 2022]


# LogisticRegression & Cross - Validation

```python
from sklearn.linear_model import LogisticRegression
LG = LogisticRegression(solver='liblinear', multi_class='ovr')
LG.fit(X_train, Y_train)

cvs_LG = cross_val_score(LG, X_train, Y_train, cv=kfold, scoring='accuracy')

y_predictLG = LG.predict(X_test)

from sklearn import metrics
# Model Accuracy, how often is the classifier correct?

# Calculate cm by calling a method named as 'confusion_matrix'
cm = confusion_matrix(Y_test, y_predictLG)
# Call a method heatmap() to plot confusion matrix
sns.heatmap(cm, annot = True)

print("Cofusion matrix: " '\n', confusion_matrix(Y_test, y_predictLG))
```

```python
print("Percent of accuracy: ", cvs_LG)
print("-----------------")
print("Accuracy: {:.2f}".format(metrics.accuracy_score(Y_test, y_predictLG)))
print("Mean Accuracy :", cvs_LG.mean(), "\nStandard Deviation :", cvs_LG.std())
  
print("-----------------")
print("Classification Report: ",classification_report(Y_test, y_predictLG))
```
